```python
import matplotlib.pyplot as plt
```

```python
df_sorted = df.sort_values(by=' Estimated Unemployment Rate (%)', ascending=False)
# Plotting the bar graph
plt.figure(figsize=(10, 8))
plt.bar(df_sorted['Region'], df_sorted[' Estimated Unemployment Rate (%)'], color='blue')
# Add labels and title
plt.xlabel('Region')
plt.ylabel(' Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate(state wise) by Bar Graph')
# Rotate the x-axis labels for readability
plt.xticks(rotation=90)
# Adjust layout to prevent labels from being cut off
plt.tight_layout()
# Display the plot
plt.show()
```

```python
df_sorted = df2.sort_values(by=' Estimated Unemployment Rate (%)', ascending=False)
# Plotting the bar graph
plt.figure(figsize=(10, 8))
plt.bar(df_sorted['Region.1'], df_sorted[' Estimated Unemployment Rate (%)'], color='blue')
# Add labels and title
plt.xlabel('Region')
plt.ylabel(' Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate(state wise) by Bar Graph')
# Rotate the x-axis labels for readability
plt.xticks(rotation=90)
# Adjust layout to prevent labels from being cut off
plt.tight_layout()
# Display the plot
plt.show()
```
